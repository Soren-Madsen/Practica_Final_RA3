## 1. Introducción a Flask

Flask es un microframework de Python para desarrollar aplicaciones web y APIs de forma sencilla y rápida. Proporciona herramientas para enrutamiento, manejo de peticiones (GET, POST, PUT, DELETE), plantillas, etc.

### 1.1. Operaciones CRUD

En el contexto de una API REST, cada operación del CRUD se asocia con un método HTTP:

| Operación | Método HTTP | Ejemplo de uso                  |
|-----------|-------------|----------------------------------|
| Create    | POST        | Crear un nuevo recurso           |
| Read      | GET         | Obtener uno o varios recursos    |
| Update    | PUT / PATCH | Actualizar un recurso existente  |
| Delete    | DELETE      | Eliminar un recurso              |

Flask permite definir rutas y asignarles los métodos HTTP correspondientes.

## Inicio de la app de Flask

```python
from flask import Flask, request, jsonify
app = flask.Flask(__name__)  #creamos la aplicación Flask.
```

Ahora abre http://127.0.0.1:5000/ en el navegador. Verás la página con el mensaje.

¿Qué está haciendo Flask? Está mapeando la URL / con la función home(). Cada vez que se accede a esa URL, se ejecuta la función y se devuelve su resultado. Este proceso se llama routing.

# Pequeña base de datos de tareas de ejemplo


```python
tasks = [
    {"id": 1, "titulo": "Aprender Flask", "completada": False},
    {"id": 2, "titulo": "Construir un API CRUD", "completada": False}
]
```

# GET - Obtener las tareas

```python
@app.route('/tasks', methods=['GET'])
def get_tasks():
    return jsonify(tasks)
```

Probar en el terminal con: 
`curl http://127.0.0.1:5000/tasks`

## POST Crear una nueva tarea

```python
@app.route('/tasks', methods=['POST'])
def create_task():
    
    new_task = {
        "id": max(t["id"] for t in tasks) + 1 if tasks else 1,
        "titulo": request.json['titulo'],
        "completada": request.json.get('completada', False)
    }
    tasks.append(new_task)
    return jsonify(new_task), 201
```

Probar en el terminal con 
`curl -X POST -H "Content-Type: application/json" -d '{"titulo": "Nueva tarea"}' http://127.0.0.1:5000/tasks`

## PUT Actualizar una tarea existente

```python
@app.route('/tasks/<int:task_id>', methods=['PUT'])
def update_task(task_id):
    task = next((t for t in tasks if t["id"] == task_id), None)
    if task is None:
        return jsonify({"error": "Tarea no encontrada"}), 404
    
    data = request.json
    task['titulo'] = data.get('titulo', task['titulo'])
    task['completada'] = data.get('completada', task['completada'])
    return jsonify(task)
```

Probar en el terminal con 
`curl -X PUT -H "Content-Type: application/json" -d '{"titulo": "Aprender Flask a fondo", "completada": true}' http://127.0.0.1:5000/tasks/1`

## DELETE Eliminar una tarea

```python
@app.route('/tasks/<int:task_id>', methods=['DELETE'])
def delete_task(task_id):
    global tasks
    task = next((t for t in tasks if t["id"] == task_id), None)
    if task is None:
        return jsonify({"error": "Tarea no encontrada"}), 404
    
    tasks = [t for t in tasks if t["id"] != task_id]
    return jsonify({"result": "Tarea eliminada"})
```

probar en el terminal con `curl -X DELETE http://127.0.0.1:5000/tasks/2`
